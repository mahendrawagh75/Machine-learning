```python
# predict the model
y_pred=l2.predict(x_test)
print(y_pred)
```

# Random over sampling

```python
import imblearn
import collections
from imblearn.over_sampling import RandomOverSampler
from collections import Counter
```

```python
ros = RandomOverSampler(random_state=0)
x_resample, y_resample = ros.fit_resample(x, y)
print(Counter(y_resample))
```

```python
Counter(x_resample.shape)
```

```python
Counter(y_resample)
```

```python
Counter(x_train.shape)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_resample,y_resample,test_size=0.2,random_state=110)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
y_sample_pred=model.predict(x_test)
y_sample_pred
```
